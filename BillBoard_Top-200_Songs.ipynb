```python
import requests
from bs4 import BeautifulSoup
```

```python
url = "https://www.billboard.com/charts/billboard-200/"
res = requests.get(url)
print(res)
```

```python
soup = BeautifulSoup(res.text, 'lxml')
```

```python
songs = soup.select("li h3")
```

```python
top_20 = [song.text[14:] [0:-5] for song in songs]
print(top_20)
```

```python
playlist = []
for i in range(0, 20):
    song = {
        "name": top_20[i]
    }
    playlist.append(song)
```

```python
print("The top 20 songs currently are:")
i = 0
for song in top_20:
    i = i + 1
    print(f"{i}) {song}")
```
